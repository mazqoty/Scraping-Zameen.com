```python
path = '/content/drive/My Drive/CSV Files/links.csv'
```

```python
#defining CSV file
csv_file = open('all_data_Khi_01.csv', 'w')
csv_writer = csv.writer(csv_file)
#defining headers
csv_writer.writerow(['Title', 'Location1', 'Type', 'Price', 'Location2',
                     'Bathrooms', 'Size', 'Purpose', 'Bedrooms'])

all_data_Khi = []
none_String = 'NaN'
df_linksFile = pd.read_csv(path, index_col = [0])
links = df_linksFile.values.tolist()

#----------------------Working with Links and opening URLs---------------------#

for i in range(len(links)):

    page_Url = str(links[i][-1])

    try:
    # Open the URL as Browser, not as python urllib
    page = urllib.request.Request(page_Url,headers={'User-Agent': 'Mozilla/5.0'})

    infile=urllib.request.urlopen(page).read()

    # Read the content as string decoded with ISO-8859-1
    data = infile.decode('ISO-8859-1')
    # Print the data to the screen
    #print(data) 

    # parses html into a soup data structure to traverse html as if it were a json data type.
    page_Soup = BeautifulSoup(infile, 'lxml')

    except Exception as e:
    page_Url = none_String

    #print(page_Soup.prettify())

    #----------------------Finding Required Elements---------------------#

    #title
    try:
    title = page_Soup.find('h1', class_='_64bb5b3b').text
    #print(title)
    except Exception as e:
    title = none_String

    #location_1 
    try:
    location_1 = page_Soup.find('div', class_='cbcd1b2b').text
    #print(location_1)
    except Exception as e:
    location_1 = none_String

    #home_Type
    try:
    home_DetailsTable = page_Soup.find_all('span', class_ ='_812aa185')
    home_Type = home_DetailsTable[0].text
    #print(home_Type)
    except Exception as e:
    home_Type = none_String

    # price
    try:
    price = page_Soup.find('div', class_ ='c4fc20ba').text
    #print(price)
    except Exception as e:
    price = none_String

    # location_2
    try:
    home_DetailsTable = page_Soup.find_all('span', class_ ='_812aa185')
    location_2 = home_DetailsTable[2].text
    #print(location_2)
    except Exception as e:
    location_2 = none_String

    # bathrooms
    try:
    home_DetailsTable = page_Soup.find_all('span', class_ ='_812aa185')
    bathrooms = home_DetailsTable[3].text
    #print(bathrooms)
    except Exception as e:
    bathrooms = none_String

    # size
    try:
    home_DetailsTable = page_Soup.find_all('span', class_ ='_812aa185')
    size = home_DetailsTable[4].text
    #print(size)
    except Exception as e:
    size = none_String

    # purpose
    try:
    home_DetailsTable = page_Soup.find_all('span', class_ ='_812aa185')
    purpose = home_DetailsTable[5].text
    #print(purpose)
    except Exception as e:
    purpose = none_String

    # bedrooms
    try:
    home_DetailsTable = page_Soup.find_all('span', class_ ='_812aa185')
    bedrooms = home_DetailsTable[6].text
    #print(bedrooms)
    except Exception as e:
    bedrooms = none_String

    #----------------------Saving Data---------------------#

    #writting to file
    csv_writer.writerow([title, location_1, home_Type, price, location_2, bathrooms, size, purpose, bedrooms])

    #problem with the following approach is that the output file of csv is too large in memory.
    """temp = {'Title': title,
          'Type': home_Type,
          'Size': size,
          'Purpose': purpose,
          'Location1':location_1,
          'Location2':location_2,
          'Bedrooms':bedrooms,
          'Bathrooms':bathrooms,
          'Price': price,
            'Url': 'dropped_off'
          }"""
    #all_data_Khi.append(temp)
    #print(all_data_Khi)
    #all_data_Khi
    """df = pd.DataFrame(all_data_Khi)
    df.to_csv('all_data_Khi_BeautifulSoup.csv', mode='a', header=False)"""
    print("Progress: ", i)

#closing CSV file
csv_file.close() 
```
